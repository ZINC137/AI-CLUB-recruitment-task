```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
train_df = pd.read_csv("atlantis_citizens_final.csv")

print("Dataset shape:", train_df.shape)
train_df.head()
train_df.info()
train_df.describe()
train_df.columns
```

```python
missing = train_df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)

```

```python
train_df["Occupation"].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Occupation", data=train_df)
plt.title("Occupation Distribution")
plt.show()
train_df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Occupation", y="Wealth_Index", data=train_df)
plt.title("Wealth Index by Occupation")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Occupation", y="House_Size_sq_ft", data=train_df)
plt.title("House Size by Occupation")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(train_df["Life_Expectancy"], bins=30, kde=True)
plt.title("Life Expectancy Distribution")
plt.show()
```

```python
categorical_cols = train_df.select_dtypes(include="object").columns
categorical_cols
for col in categorical_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(x=col, data=train_df)
    plt.title(f"{col} Distribution")
    plt.show()
```

```python
for col in categorical_cols:
    print("\n", col)
    print(train_df[col].value_counts().head())
```
